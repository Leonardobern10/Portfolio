```python
#Regressão linear
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from yellowbrick.regressor import ResidualsPlot
```

```python
#Regressão linear simples
d = {'Idade':[16, 25, 39, 45, 49, 64, 70, 29, 57, 22],
         'Pressão_Sanguínea':[109, 122, 143, 132, 199, 185, 199, 130, 175, 118]}
dados = pd.DataFrame(d)
dados
```

```python
#Regressão linear simples

#Idade é a variavel independente (x)
#Pressão sanguinea é a variavel dependente (y)
#Fazendo a separação
x = dados.iloc[:, 0].values
y = dados.iloc[:, 1].values
```

```python
#Regressão linear simples

#Calculando a correlação entre x e y
correlação = np.corrcoef(x, y)
correlação
```

```python
#Regressão linear simples

x = x.reshape(-1, 1)
modelo = LinearRegression()
modelo.fit(x, y)
```

```python
#regressão linear simples

# é o valor de b, o intercepto em y
#valor onde a reta corta o eixo y
modelo.intercept_
```

```python
#regressão linear simples

#Inclinação da reta de regressão
modelo.coef_
```

```python
#coeficiente de determinação r^2
modelo.score(x, y)
```

```python
#regressão linear simples

plt.scatter(x, y)
plt.plot(x, modelo.predict(x), color = 'red')
```

```python
#regressão linear simples

#Qual a pressão sanguinea de um adulto com 35 anos?
# interceptação + inclinação * pressão - pela formula: y = mx + b
modelo.intercept_ + modelo.coef_ * 35
```

```python
#regressão linear simples

#Mesma previsão acima, porem utilizando o sklearn
modelo.predict([[35]])
```

```python
#regressão linear simples

vizualizador = ResidualsPlot(modelo)
vizualizador.fit(x, y)
vizualizador.poof()
```
