```python
import pandas as pd
import seaborn as sns
```

```python
dados = pd.read_csv('/content/Salaries.csv')
dados
```

```python
del dados ['Unnamed: 0']
```

```python
dados
```

```python
dados.columns = ['Ranking', 'Disciplina', 'Phd(Anos)', 'Experiência(Anos)', 'Sexo', 'Salário']
dados
```

```python
for x in dados.columns:
  print(x, ':', dados[x].isnull().sum())
```

```python
ranking = dados.groupby(['Ranking']).size()
ranking
```

```python
ranking.plot.bar()
```

```python
disciplina = dados.groupby(['Disciplina']).size()
disciplina
```

```python
disciplina.plot.bar()
```

```python
phd = dados.groupby(['Phd(Anos)']).size()
phd
```

```python
sns.boxplot(dados['Phd(Anos)']).set_title('Phd')
```

```python
experiencia = dados.groupby(['Experiência(Anos)']).size()
experiencia
```
