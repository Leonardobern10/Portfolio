```python
#Amostragem sitemática
''' Cria variaveis correpondentes ao total de instâncias e a quantidade de amostras. Baseado nisso, cria a variavel k com a função ceil (p/a). 
A partir daí com numpy, escolhe um valor aleatorio em um range definido, e desse valor é criado um laço para que seja adcionado em uma variavel
definida como lista, todos os valores que correspodem a uma adição do valor k ao valor escolhido aleatoriamente por numpy, para obter a quantidade
de amostras pré-selecionadas. Por exemplo: se eu quero 4 amostras, 2(primeiro numero escolhido aleatoriamente) + 10(k) = 12, 12+10(k)=22, 22+10(k)=32,
32+10(k)=42. 
Então temos essas 4 amostras aplicadas no dataframe para que as instancias correspondentes sejam escolhidas e armazenadas em uma variavel, 
separando-as do dataframe.

'''
populaçao = 50
amostra = 5
k = ceil(populaçao/amostra)
print(k)
```

```python
#Amostragem estratificada
r = np.random.randint(low = 1, high = k+1, size = 1)
print(r)
```

```python
#Amostragem estratificada
acumulador = r[0]
sorteados = []
for i in range(amostra):
  #print(acumulador)
  sorteados.append(acumulador)
  acumulador += k
print(sorteados)
```

```python
#Amostragem estratificada
len(sorteados)
```

```python
#Distribuição normal (Apesar disso os dados não estão normalmente distribuidos como o teste abaixo comprova)

#Uma vez que o Teorema central do limite diz que amostras com mais de 30 elementos são consideradas normalmente distribuídas, e que a medida
#que esse numero aumenta, mais normalmente distribuidas elas se encontram.

#Temos média = 1857.66 e desvio padrão = std = 1163.2664181484122

media = np.mean(vendas_unid)
std = np.std(vendas_unid, ddof = 1)
mediana = np.median(vendas_unid)
estatisticas = stats.describe(vendas_unid)
```

```python
#Distribuição normal (Apesar disso os dados não estão normalmente distribuidos como o teste abaixo comprova)

#Qual a probabilidade de vender menos que 500.000 por unidade?
norm.cdf(500, media, std )
```

```python
#Distribuição normal (Apesar disso os dados não estão normalmente distribuidos como o teste abaixo comprova)

#Qual a probabilidade de vender mais que 1.500.00 por unidade?
norm.sf(1500, media, std) #ou 1 - norm.cdf(1500, media, std)
```

```python
#Distribuiçao normal (Apesar disso os dados não estão normalmente distribuidos como o teste abaixo comprova)

#Qula a probabilidade de vender menos que 500.000 ou mais que 1.500.000 por unidade?
norm.cdf(500, media, std) + norm.sf(1500, media, std)
```

```python
#Distribuição normal (Apesar disso os dados não estão normalmente distribuidos como o teste abaixo comprova)

#Qual a probabilidade de vender entre 1.000.000 e 1.500.000 por unidade?
norm.sf(1000, media, std) - norm.sf(1500, media, std)
```

```python
#Teste de normalidade
#A hipotese nula é por padrao que os dados estão normalmente distribuidos, dados isso:
#Se p-value >= 0,05, a hipotese nula está mantida
#Se p-value < 0,05, a hipotese nula está descartada
from scipy import stats
from scipy.stats import skewnorm, norm
import matplotlib.pyplot as plt
```
